Temel Düzey

En yüksek maaşı alan çalışanı bul ve adını yazdır.

Performans skoru 5 olan çalışanların sayısını bul.

#### En yüksek maaşı alan çalışanı bulma

```python
df = pd.read_csv('/content/8-apply_function_data.csv')
df
df.max()
```

```python
df.describe()
```

#### Departman bazında ortalama maaşı hesapla.

```python
grouped_department = df.groupby('Department')
grouped_department.max()
```

#### Performans skoru 5 olan çalışanların sayısını bul.

```python
score_5 = df[df["Performance_Score"] == 5].count()
score_5
```

#### Deneyim yılı 10’dan fazla olan çalışanların ortalama performans skorunu hesapla.





```python
big_10 = df[df["Experience"] >10]
big_10
big_10["Performance_Score"].mean()
```

#### Maaşı departman ortalamasının üzerinde olan çalışanların listesini çıkar.

```python
# 1. Her departmanın ortalama maaşını hesapla
salary_mean = df.groupby('Department')['Salary'].mean()
# 2. Her satır için karşılaştırma yaparak filtrele
above_avg = df[df.apply(lambda row: row['Salary'] > salary_mean[row['Department']], axis=1)]
# 3. Sonucu göster
print(above_avg)
```

#### En yüksek ortalama maaşa sahip departmanı bul

```python
# 1. Departman ortalama maaşlarını hesapla
avg_salaries = df.groupby('Department')['Salary'].mean()

# 2. En yüksek ortalama maaşa sahip departmanı bul
max_dept = avg_salaries.idxmax()   # Departman ismi
max_salary = avg_salaries.max()    # Maaş değeri

print("En yüksek ortalama maaş:", max_salary)
print("Departman:", max_dept)
```

Performans skoruna göre çalışanları “Düşük”, “Orta”, “Yüksek” şeklinde kategorilere ayıran yeni bir sütun ekle. (ör. 1–2 düşük, 3–4 orta, 5 yüksek)

Departmana göre toplam maaş maliyetini hesapla ve en pahalı departmanı bul.

Deneyim yılı ve performans skorunu birlikte değerlendirerek, maaşı en “beklenmedik” yüksek olan çalışanı belirle (ör. az deneyim ve düşük performansa rağmen yüksek maaş).
