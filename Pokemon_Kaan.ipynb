```python
#corelation map - featurelar arası ilgiyi anlamak için 
#corelation 1 ise doğru orantılıdır
#corelation -1 ise ters orantılıdır
#corelation sıfıra yaklaştıkça ilişki az demektir

# sadece sayısal sütunları al
corr = df.select_dtypes(include='number').corr()

# korelasyon ısı haritası
f, ax = plt.subplots(figsize=(13, 13))
sns.heatmap(corr, annot=True, linewidths=0.5, fmt=".2f", ax=ax)
plt.show()
```
